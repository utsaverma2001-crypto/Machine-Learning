```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,20)
```

```python
y=x*x
z=x+y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,z)
```

```python
plt.plot(x,y,x,x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Example')
```

```python
plt.subplot(3,2,1)
plt.plot(x,x)
plt.subplot(3,2,2)
plt.plot(x,y)
plt.subplot(3,2,3)
plt.plot(y,x)
plt.subplot(3,2,4)
plt.plot(x,z)
plt.subplot(3,2,5)
plt.plot(y,z)
plt.subplot(3,2,6)
plt.plot(z,z)
plt.tight_layout()
```

```python
plt.subplot(3,2,1)  #diffrent types of plots
plt.plot(x,x)
plt.subplot(3,2,2)
plt.scatter(x,y)
plt.subplot(3,2,3)
plt.hist(y,x)
plt.subplot(3,2,4)
plt.fill(x,z)
plt.subplot(3,2,5)
plt.bar(y,z)
plt.subplot(3,2,6)
plt.hist(z,z)
plt.tight_layout()
```

```python
plt.polar(x,y)
```

```python
plt.plot(x,y,label="freq") #legend
plt.legend(loc=2) #loc ranges from 1->10
```

```python
fig=plt.figure()  #object oriented plots
axes=fig.add_axes([0.1,0.1,1.0,1.0])
axes.plot(x,y)
```

```python
fig=plt.figure(figsize=(4,4))  
axes=fig.add_axes([0.1,0.1,1.0,1.0])
axes1=fig.add_axes([0.3,0.3,0.6,0.6])
```

```python
fig=plt.figure()  
axes=fig.add_axes([0.1,0.1,0.6,0.6])
axes.plot(x,z)
axes1=fig.add_axes([0.7,0.7,0.6,0.6])
axes1.plot(z,x)
```

```python
fig,axes=plt.subplots(1,2) #styling and labeling
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')
axes[0].plot(x,y,color='r',linewidth=2)
axes[1].plot(y,x,alpha=0.9,linestyle='--')
plt.tight_layout()
```

```python
fig=plt.figure()  
axes=fig.add_axes([0.1,0.1,0.6,0.6])
axes.plot(x,z,marker='o')

```

```python
fig=plt.figure()  #markups
axes=fig.add_axes([0.1,0.1,0.6,0.6])
axes.plot(x,z,marker='1',markersize=10,markerfacecolor='green')

```

```python
fig=plt.figure()  #setting limits
axes=fig.add_axes([0.1,0.1,0.6,0.6])
axes.plot(x,y)
axes.set_xlim([0,8])
axes.set_ylim([0,60])
```
