```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data=sns.load_dataset('tips') #loading data of .csv file
```

```python
data.head(10)
```

```python
 # pointplots
sns.displot(data['size'])
```

```python
sns.rugplot(data['total_bill'])
```

```python
sns.kdeplot(data['tip'])
```

```python
#distribution plots
```

```python
sns.residplot(x='total_bill',y='tip',data=data)
```

```python
sns.jointplot(x='total_bill',y='tip',data=data)
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='hex',color="#4CB391")
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='reg')
```

```python
sns.pairplot(data,hue="time") #used to analyse all data at once
```

```python
sns.pairplot(data)
```

```python
 # Categorial plots
```

```python
sns.countplot(x='day',data=data)
```

```python
sns.pointplot(x='day',y='total_bill',data=data)
```

```python
sns.catplot(x='day',y='tip',data=data)
```

```python
sns.factorplot(x='day',y='tip',data=data,kind='box')
```

```python
sns.swarmplot(x='sex',y='tip',data=data)
```

```python
sns.stripplot(x='day',y='total_bill',data=data,hue='sex')
```
